### Drop all social; create boolean feature for website

- Websites are String objects
- Nulls are float objects

```python
def new_col_boolean(dfName, colName, valType):
    new_col = []
    for val in dfName[colName]:
        if (type(val)) == valType:
            new_col.append(1)
        else:
            new_col.append(0)
    return new_col
```

```python
market_df['has_website'] = new_col_boolean(market_df, 'Website', str)
```

```python
cols = ['Website', 'has_website']
market_df[cols].head()
```

```python
market_df['has_website'].value_counts()
```

```python
market_df.drop(columns=social_cols, axis=1, inplace=True)
market_df.head()
```

## Binary Product Columns

```python
market_df['Has Market'] = [1]*market_df.shape[0]
```

```python
product_cols = ['Organic', 'Bakedgoods', 'Cheese', 'Crafts', 'Flowers', 'Eggs', 
                'Seafood', 'Herbs', 'Vegetables', 'Honey', 'Jams', 'Maple', 'Meat', 
                'Nursery', 'Nuts', 'Plants', 'Poultry', 'Prepared', 'Soap', 'Trees', 
                'Wine', 'Coffee', 'Beans', 'Fruits', 'Grains', 'Juices', 
                'Mushrooms', 'PetFood', 'Tofu', 'WildHarvested']
```

```python
for product in product_cols:
    try:
        market_df[product] = market_df[product].replace(to_replace=['Y', 'N'], value=[1,0])
    except:
        continue
```

```python
market_df.head()
```

# Categories

```python
products_df = pd.DataFrame()
```

```python
new_products = {'Produce': ['Vegetables', 'Fruits', 'Mushrooms', 'WildHarvested', 'Beans', 'Grains', 'Herbs', 'Organic', 'Nuts'],
              'Confectionary': ['Bakedgoods', 'Honey', 'Jams', 'Maple', 'Juices'],
              'Household': ['Crafts', 'Soap', 'PetFood'],
              'Dairy': ['Cheese', 'Eggs', 'Tofu'],
              'Plants': ['Flowers', 'Nursery', 'Plants', 'Trees'],
              'Meat': ['Seafood','Meat', 'Poultry'],
              'Other': ['Wine', 'Prepared', 'Coffee']}
```

```python
for product in new_products.keys():
    try:
        products_df[product] = market_df.loc[:,new_products[product]].sum(1)
    except:
        print(product)
        
products_df['product_count'] = products_df.sum(1)
```

```python
products_df.head()
```

```python
market_df = market_df.drop(columns=product_cols)
market_df = pd.concat([market_df, products_df], axis=1)
```

```python
market_df = market_df.loc[market_df['product_count'] != 0.0]
market_df.head()
```

## Payment Methods

```python
payment_methods = ['Credit', 'WIC', 'WICcash', 'SFMNP', 'SNAP']
for payment_method in payment_methods:
    try:
        market_df[payment_method] = market_df[payment_method].replace(to_replace=['Y', 'N'], value=[1,0])
    except:
        continue
market_df['alt_pay_methods'] = market_df.loc[:, 'WIC':'SNAP'].sum(1)
```
